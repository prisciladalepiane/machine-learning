# Probabilidade

Probabilidade é uma medida numérica que descreve a chance de um evento ocorrer. Em termos mais simples, é a chance ou a possibilidade de que algo aconteça.

Em contextos matemáticos e estatísticos, a probabilidade é representada por um número entre 0 e 1, onde 0 indica impossibilidade absoluta do evento ocorrer, e 1 indica certeza absoluta de que o evento ocorrerá.

Seja E um evento e S o espaço amostral (Conjunto de todos os dados possíveis), a probabilidade do evento E, denotado por P(E) é dado por:

$$ P(E) = \frac{n(E)}{n(S)} $$

onde:

- $n(E)$ é o número de resultados favoráveis ao evento E.
- $n(S)$ é o número total de resultados possíveis.


Um **experimento aleatório** é um processo ou procedimento que, quando realizado, pode gerar diferentes resultados, e não podemos prever com certeza qual resultado específico ocorrerá.
Exemplos comuns de experimentos aleatórios incluem lançar um dado, lançar uma moeda, escolher uma carta de um baralho, entre outros.

O **espaço amostral** de um experimento aleatório é o conjunto de todos os resultados possíveis desse experimento. É importante notar que cada resultado possível deve ser único e mutuamente exclusivo.
Por exemplo, ao lançar um dado de seis faces, o espaço amostral é o conjunto {1, 2, 3, 4, 5, 6}, pois estes são todos os resultados possíveis desse experimento.

Quando realizamos um experimento aleatório, o espaço amostral consiste em todos os resultados possíveis desse experimento e os eventos são conjuntos de resultados específicos dentro desse espaço amostral, dos quais estamos interessados em considerar a ocorrência.

A teoria da probabilidade nos permite quantificar a probabilidade de diferentes eventos ocorrerem com base nas propriedades do espaço amostral e nas regras probabilísticas.

## Tipos de probabilidade

### Probabilidade Clássica

Também conhecida como probabilidade teórica, é aplicável a experimentos em que todos os resultados são igualmente prováveis.

A probabilidade de um evento é determinada dividindo-se o número de resultados favoráveis pelo número total de resultados possíveis.

Por exemplo, lançar uma moeda justa ou um dado não viciado.

### Probabilidade Empírica (ou Frequencial)
Baseia-se em observações ou experimentos repetidos. A probabilidade de um evento é estimada pela frequência relativa com que esse evento ocorre em um grande número de experimentos idênticos ou similares.

Por exemplo, se lançarmos um dado muitas vezes e contarmos quantas vezes cada número aparece, a probabilidade de cada número será aproximadamente a mesma.

### Probabilidade Subjetiva

Refere-se à probabilidade baseada em julgamento pessoal ou opinião, muitas vezes sem a necessidade de dados empíricos. Ela expressa a incerteza de um indivíduo sobre a ocorrência de um evento.

Por exemplo, a probabilidade de chuva amanhã pode ser expressa subjetivamente por uma pessoa com base em sua experiência, conhecimento do clima e outros fatores.


Cada tipo de probabilidade tem suas próprias aplicações e limitações, e a escolha do tipo de probabilidade depende do contexto específico do problema e dos dados disponíveis.

Em muitas situações práticas, uma combinação desses tipos de probabilidade pode ser usada para fazer previsões ou tomar decisões.


## O Que São Variáveis Aleatórias?

Uma variável aleatória é uma função que associa um número real a cada resultado possível de um experimento aleatório. Em outras palavras, ela atribui valores numéricos aos resultados de um experimento aleatório.

As variáveis aleatórias são fundamentais na teoria da probabilidade e na estatística, pois permitem a modelagem matemática de fenômenos aleatórios e a análise de seus comportamentos.

Existem dois tipos principais de variáveis aleatórias:

### Variáveis Aleatórias Discretas
Uma variável aleatória discreta assume um conjunto finito ou infinito contável de valores.

Por exemplo, o número de caras em 3 lançamentos de uma moeda é uma variável aleatória discreta, pois pode assumir apenas valores inteiros (0, 1, 2 ou 3) e não pode assumir valores entre esses números.

### Variáveis Aleatórias Contínuas
Uma variável aleatória contínua pode assumir qualquer valor dentro de um intervalo específico de números reais.

Por exemplo, a altura de uma pessoa é uma variável aleatória contínua, pois pode assumir qualquer valor dentro de um intervalo específico (por exemplo, de 1 metro a 2 metros).

As variáveis aleatórias são usadas para representar características de interesse em uma variedade de situações, desde resultados de jogos até medidas em experimentos científicos.

Elas são uma ferramenta poderosa para modelar e analisar a incerteza e a variabilidade em uma ampla gama de contextos.

### Valor médio de uma variável aleatória discreta

O valor médio de uma variável aleatória discreta, também conhecido como **esperança matemática** ou valor esperado, é uma medida de tendência central que representa o valor médio que a variável assume ponderado pelas suas probabilidades de ocorrência.

Para uma variável aleatória discreta $X$ com conjunto de valores possíveis $\{x_1, x_2, ... ,x_n\}$ e suas probabilidades $\{p_1, p_2, ... ,p_n\}$, o valor médio $E(X)$ é calculado como:

$$ E(X) = \sum_{i=1}^{n}x_i.p_i$$

### Desvio Padrão e Variância de Uma Variável Aleatória Discreta

O desvio padrão e a variância são medidas de dispersão que indicam o quão distantes os valores de uma variável aleatória estão da média.

**Variância**

A variância de uma variável aleatória discreta mede a dispersão dos seus valores em relação à média.

Se a variância é alta, isso significa que os valores estão mais espalhados em torno da média; se é baixa, os valores tendem a estar mais próximos da média.

**Desvio Padrão**

O desvio padrão é simplesmente a raiz quadrada da variância. Ele tem a mesma interpretação que a variância, mas na mesma escala dos dados originais.

Ou seja, é uma medida de dispersão que nos dá uma ideia da magnitude do desvio dos valores em relação à média.

A variância e o desvio padrão nos ajudam a entender o quanto os valores de uma variável aleatória se afastam da média, o que é fundamental para avaliar a variabilidade e a incerteza nos dados.

Quanto maior o desvio padrão ou a variância, maior a dispersão dos valores em relação à média, e vice-versa.

# Distribuições de Probabilidade 

Distribuições de probabilidade são modelos matemáticos que descrevem o comportamento aleatório de uma variável ou conjunto de variáveis. Elas atribuem probabilidades a diferentes resultados possíveis de um experimento aleatório ou de um fenômeno aleatório. Essas distribuições são fundamentais na teoria da probabilidade e na estatística, pois fornecem uma maneira de entender e analisar a incerteza e a variabilidade em uma ampla gama de contextos.

Elas são frequentemente utilizadas em situações onde os resultados são influenciados por fatores aleatórios, como em experimentos científicos, previsões de mercado, processos industriais, entre outros. Existem diferentes tipos de distribuições de probabilidade, cada uma com suas próprias propriedades e aplicações. 

## Distribuições de probabilidade de variáveis aleatórias discretas 

As distribuições de probabilidade de variáveis aleatórias discretas são ferramentas de análise que nos ajudam a entender e quantificar a incerteza em processos ou experimentos que resultam em resultados contáveis. Elas fornecem uma maneira de calcular a probabilidade de ocorrência de cada possível resultado de uma variável aleatória discreta.

As distribuições de probabilidade de variáveis aleatórias discretas descrevem a probabilidade de ocorrência de cada valor de uma variável aleatória que assume valores discretos. 

Aqui estão algumas maneiras pelas quais essas distribuições podem ser usadas como ferramentas de análise:

- **Descrever a distribuição**: Elas nos permitem descrever a distribuição dos resultados possíveis de um experimento ou processo, especificando a probabilidade de cada resultado.

- **Calcular probabilidades**: Podemos usar as distribuições para calcular a probabilidade de diferentes eventos, como a probabilidade de um resultado específico ou a probabilidade de um resultado estar dentro de um intervalo específico.

- **Medir tendência central e dispersão**: As distribuições permitem calcular medidas como a média (ou esperança matemática), que indica a tendência central, e a variância ou desvio padrão, que indicam a dispersão ou variabilidade dos resultados.

- **Realizar inferências estatísticas**: Podemos usar essas distribuições para realizar testes de hipóteses e construir intervalos de confiança, ajudando a tomar decisões com base em dados.

- **Modelagem e simulação**: As distribuições de probabilidade são frequentemente usadas para modelar fenômenos reais em diversas áreas, como finanças, engenharia e ciências sociais. Elas também são usadas em simulações computacionais para gerar dados que imitam o comportamento de sistemas reais.

Algumas das principais distribuições de probabilidade são:

- **Distribuição Uniforme:**  Todos os resultados possíveis têm a mesma probabilidade de ocorrência. Por exemplo, ao lançar uma moeda justa, a distribuição de probabilidade para cara e coroa seria uniforme.
- **Distribuição Binomial**: Modela o número de sucessos em uma série de tentativas independentes, onde cada tentativa tem apenas dois resultados possíveis (por exemplo, sucesso ou fracasso), com uma probabilidade constante de sucesso em cada tentativa.
- **Distribuição Normal (Gaussiana)**: Uma das distribuições mais importantes e amplamente utilizadas, que descreve muitos fenômenos naturais. É caracterizada por uma forma de sino e é completamente determinada por sua média e desvio padrão.
- **Distribuição de Poisson**: Modela a ocorrência de eventos em um intervalo de tempo ou espaço, quando os eventos ocorrem de forma independente e a uma taxa média constante.
- **Distribuição Exponencial**: Descreve o tempo entre eventos sucessivos em um processo de Poisson, ou seja, é usada para modelar o tempo de espera até que ocorra um evento.
Esses são apenas alguns exemplos das muitas distribuições de probabilidade que existem.


```python
# Imports
import scipy
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
from scipy.stats import poisson
from scipy.stats import geom
from scipy.stats import nbinom
from scipy.stats import hypergeom
```

## Distribuições Discretas

### Distribuição Uniforme

Todos os resultados possíveis têm a mesma probabilidade de ocorrência. Por exemplo, ao lançar uma moeda justa, a distribuição de probabilidade para cara e coroa seria uniforme.


```python
# Define o número de lançamentos
n_lancamentos = 1000

# Gera n_lancamentos de números aleatórios entre 1 e 6
lancamentos = np.random.randint(1, 7, size = n_lancamentos)

# Calcula e imprime a frequência de cada resultado
for i in range(1, 7):
    frequencia = np.sum(lancamentos == i)
    print(f"Face {i}: {frequencia} vezes ({frequencia / n_lancamentos * 100:.2f}%)")
```

```python
# Calcula a frequência de cada resultado
frequencias = [np.sum(lancamentos == i) for i in range(1, 7)]
```

```python
# Cria um gráfico de barras para mostrar a frequência de cada face
plt.figure(figsize = (9, 4))
faces = range(1, 7)
plt.bar(faces, frequencias, color = 'purple')
plt.xlabel('Face do Dado')
plt.ylabel('Frequência')
plt.title('Frequência dos Resultados de 1000 Lançamentos de um Dado')
plt.xticks(faces)
plt.show()
```

### Distribuição Binomial

Descreve o número de sucessos em n ensaios independentes de Bernoulli (experimentos com dois possíveis resultados, como sucesso e fracasso) com a mesma probabilidade de sucesso p. 
Por exemplo, o número de caras em 10 lançamentos de uma moeda.
O exemplo abaixo simula um cenário comum de lançamento de moeda, onde estamos interessados em encontrar a probabilidade de obter um determinado número de caras (sucesso) em um número fixo de lançamentos.

Imagine que você lança uma moeda justa 10 vezes e quer saber a probabilidade de obter exatamente 5 caras. A moeda justa significa que a probabilidade de cara (sucesso) é de 0,5 em cada lançamento.

```python
# Define os parâmetros da distribuição binomial

# Número de ensaios (lançamentos da moeda)
n = 12

# Probabilidade de sucesso (cara) em cada ensaio
p = 0.5  

# Calcula a probabilidade de obter exatamente 5 sucessos (caras)
probabilidade = binom.pmf(5, n, p)

print(f"\nA probabilidade de obter exatamente 5 caras em 12 lançamentos é de {probabilidade * 100:.2f}%")
```

```python
# Gera um gráfico da distribuição binomial
plt.figure(figsize = (10, 4))
x = range(n + 1)
y = binom.pmf(x, n, p)
plt.bar(x, y, color = 'purple')
plt.xlabel('Número de Caras')
plt.ylabel('Probabilidade')
plt.title('Distribuição Binomial para 12 Lançamentos de uma Moeda Justa')
plt.xticks(x)
plt.show()
```

### Distribuição de Poisson

Usada para modelar o número de vezes que um evento acontece em um intervalo de tempo ou espaço fixo, quando esses eventos ocorrem com uma taxa média constante e independente do tempo desde o último evento. Por exemplo, o número de chamadas recebidas por um call center em uma hora.

Suponha que queremos modelar o número de chamadas que um call center recebe em uma hora. Se a taxa média de chamadas é de 10 chamadas por hora, podemos usar a distribuição de Poisson para calcular a probabilidade de receber um determinado número de chamadas, digamos 15, em uma hora.

```python
# Define a taxa média de chamadas por hora
lambda_ = 10  # 'lambda' é a taxa média de eventos (chamadas) por intervalo (hora)

# Calcula a probabilidade de receber exatamente 15 chamadas em uma hora
probabilidade = poisson.pmf(15, lambda_)

print(f"\nA probabilidade de receber exatamente 15 chamadas em uma hora é de {probabilidade * 100:.2f}%")
```

```python
# probabilidade de receber pelo menos 15 chamadas em uma hora
probabilidade2 = 1 - poisson.cdf(14, lambda_)

print(f"\nA probabilidade de receber pelo menos 15 chamadas em uma hora é de {probabilidade2 * 100:.2f}%")
```

```python
# Gera um gráfico da distribuição de Poisson
plt.figure(figsize = (10, 4))
x = range(0, 25)  # Define um intervalo de valores possíveis para o número de chamadas
y = poisson.pmf(x, lambda_)
plt.bar(x, y, color = 'purple')
plt.xlabel('Número de Chamadas')
plt.ylabel('Probabilidade')
plt.title('Distribuição de Poisson para Chamadas Recebidas em uma Hora')
plt.xticks(x)
plt.show()
```

```python
# Gera um gráfico da distribuição de Poisson
plt.figure(figsize = (10, 4))
x = range(0, 25)  # Define um intervalo de valores possíveis para o número de chamadas
y = poisson.cdf(x, lambda_)
plt.bar(x, y, color = 'purple')
plt.xlabel('Número de Chamadas')
plt.ylabel('Probabilidade')
plt.title('Distribuição Acumulada de Poisson para Chamadas Recebidas em uma Hora')
plt.xticks(x)
plt.show()
```

### Distribuição Geométrica

Modela o número de ensaios de Bernoulli necessários para obter o primeiro sucesso. Por exemplo, o número de lançamentos de uma moeda até que saia cara pela primeira vez.

A distribuição geométrica modela o número de ensaios de Bernoulli (experimentos com dois possíveis resultados, sucesso ou fracasso) necessários para obter o primeiro sucesso, assumindo que a probabilidade de sucesso em cada ensaio é constante.

Por exemplo, consideremos o lançamento repetido de uma moeda justa, onde estamos interessados na probabilidade de que o primeiro "cara" (sucesso) ocorra no quinto lançamento. A probabilidade de sucesso (obter "cara") em cada lançamento é de 0.5.

Neste código, `geom.pmf(k, p)` calcula a função de massa de probabilidade para a distribuição geométrica, que é a probabilidade de que o primeiro sucesso ocorra no k-ésimo ensaio, dado que a probabilidade de sucesso em cada ensaio é p. O resultado nos dá a probabilidade de que o primeiro "cara" apareça especificamente no quinto lançamento de uma moeda.

```python
p = 0.5  # Probabilidade de cara

# Calcula a probabilidade de o primeiro sucesso ocorrer no 5º lançamento
probabilidade = geom.pmf(5, p)

print(f"\nA probabilidade de o primeiro 'cara' ocorrer no 5º lançamento é de {probabilidade * 100:.2f}%")
```

```python
# Gera um gráfico da distribuição geométrica
plt.figure(figsize = (10, 4))
x = range(1, 11)  # Define um intervalo de valores possíveis para o número de lançamentos até o primeiro sucesso
y = geom.pmf(x, p)
plt.bar(x, y, color = 'purple')
plt.xlabel('Número de Lançamentos Até o Primeiro Sucesso')
plt.ylabel('Probabilidade')
plt.title('Distribuição Geométrica para o Primeiro Sucesso em Lançamentos de uma Moeda')
plt.xticks(x)
plt.show()
```

```python
# Gera um gráfico da distribuição geométrica
plt.figure(figsize = (10, 4))
x = range(1, 11)  # Define um intervalo de valores possíveis para o número de lançamentos até o primeiro sucesso
y = geom.cdf(x, p)
plt.bar(x, y, color = 'purple')
plt.xlabel('Número de Lançamentos Até o Primeiro Sucesso')
plt.ylabel('Probabilidade')
plt.title('Distribuição Acumulada Geométrica')
plt.xticks(x)
plt.show()
```

### Distribuição Binomial Negativa

Generalização da distribuição geométrica, essa distribuição modela o número de ensaios de Bernoulli necessários para alcançar um número específico de sucessos.

Por exemplo, o número de lançamentos de uma moeda até que saia cara três vezes.

A distribuição binomial negativa modela o número de ensaios de Bernoulli necessários para alcançar um determinado número de sucessos, com a probabilidade de sucesso sendo constante em cada ensaio.

Por exemplo, considere um experimento onde você está lançando uma moeda justa e quer saber a probabilidade de que seja necessário realizar 10 lançamentos para obter 3 caras (sucessos). A probabilidade de cara (sucesso) em cada lançamento é de 0.5.

```python
r = 3  # Número de sucessos desejados

p = 0.5  # Probabilidade de sucesso em cada ensaio

# Probabilidade de que sejam necessários 10 lançamentos para obter 3 sucessos
prob = nbinom.pmf(10-r, r, p)

print(f"\nA probabilidade de que sejam necessários 10 lançamentos para obter 3 caras é de {prob * 100:.2f}%")
```

Neste exemplo, usamos nbinom.pmf(k, r, p) para calcular a função de massa de probabilidade da distribuição binomial negativa. Aqui, k é o número total de ensaios menos o número de sucessos (10 - 3 neste caso, porque queremos saber a probabilidade de que os três sucessos ocorram nos primeiros 10 ensaios), r é o número de sucessos desejados, e p é a probabilidade de sucesso em cada ensaio. O resultado representa a probabilidade de precisar de exatamente 10 lançamentos para obter 3 sucessos.

### Distribuição Hipergeométrica

Modela o número de sucessos em amostras retiradas sem reposição de uma população finita que contém um número fixo de sucessos e fracassos. Por exemplo, o número de cartas vermelhas retiradas em uma mão de 5 cartas de um baralho padrão de 52 cartas.

A distribuição hipergeométrica descreve a probabilidade de obter um determinado número de sucessos em amostras retiradas sem reposição de uma população finita que consiste em dois tipos de objetos, sucessos e fracassos.

Imagine um cenário onde temos um baralho de 52 cartas, das quais 26 são vermelhas (sucessos) e 26 são pretas (fracassos). Você quer saber a probabilidade de retirar exatamente 5 cartas vermelhas ao escolher 10 cartas aleatoriamente do baralho, sem reposição.

```python
# Define os parâmetros da distribuição hipergeométrica

# Tamanho total da população (baralho de cartas)
M = 52  

# Total de sucessos na população (cartas vermelhas)
n = 26  

# Tamanho da amostra (número de cartas retiradas)
N = 10  

# Calcula a probabilidade de obter exatamente 5 sucessos (cartas vermelhas) na amostra
prob = hypergeom.pmf(5, M, n, N)

print(f"\nA probabilidade de retirar exatamente 5 cartas vermelhas ao escolher 10 cartas é de {prob * 100:.2f}%")
```

Neste código, hypergeom.pmf(k, M, n, N) calcula a função de massa de probabilidade da distribuição hipergeométrica, onde k é o número de sucessos desejados na amostra (5 cartas vermelhas), M é o tamanho total da população (52 cartas), n é o número total de sucessos na população (26 cartas vermelhas), e N é o tamanho da amostra (10 cartas retiradas). O resultado nos dá a probabilidade de retirar exatamente 5 cartas vermelhas de um baralho de 52 cartas ao escolher 10 cartas sem reposição.

```python
# Plot
plt.figure(figsize = (10, 4))
x = range(0, 11)  # Define um intervalo de valores possíveis para o número de sucessos na amostra
y = hypergeom.pmf(x, M, n, N)
plt.bar(x, y, color = 'purple')
plt.xlabel('Número de cartas vermelhas na amostra')
plt.ylabel('Probabilidade')
plt.title('Distribuição Hipergeométrica para escolha de 10 cartas de um baralho')
plt.xticks(x)
plt.show()
```

No entanto, para garantir a alta qualidade do produto final, a empresa define uma política de que o software deve passar por 3 testes de QA consecutivos antes de ser considerado pronto para o lançamento.
