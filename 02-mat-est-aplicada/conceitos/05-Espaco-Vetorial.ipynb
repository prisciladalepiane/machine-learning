# Vetores e √Ålgebra Linear no Espa√ßo Vetorial

## Axiomas de Opera√ß√µes com Vetores no Espa√ßo Vetorial 

Um espa√ßo vetorial, como o pr√≥prio nome indica, √© o espa√ßo para vetores, que define duas opera√ß√µes, **adi√ß√£o** e **multiplica√ß√£o por escalares**, sujeitas aos axiomas abaixo:

1. A soma de ${u}$ e ${v}$, indicado por ${u}+{v},$ est√° em $V$
2. ${u}+{v}={v}+{u}$
3. $({u}+{v})+{w}={u}+({v}+{w})$
4. Existe um vetor nulo $0$ em $V$ tal que ${u}+{0}={u}$
5. Para cada ${u}$ em $V$, h√° um vetor $-{u}$ in $V$ tal que ${u}+(-{u})={0}$
6. A multipplica√ß√£o do escalar ${u}$ por $c,$ indicado por $c {u},$ est√° em $V$
7. $c({u}+{v})=c {u}+c {v}$
8. $(c+d) {u}=c {u}+d {u}$
9. $c(d {u})=(c d) {u}$
10. $1 {u}={u}$

# Subespa√ßo em √Ålgebra Linear

Um subespa√ßo √© um dos conceitos mais importantes em √°lgebra linear, felizmente nada realmente misterioso.

Um subespa√ßo reside em um espa√ßo vetorial $V$, podemos denotar como $H$. Apenas duas propriedades precisam ser verificadas:

1. $H$ tem vetor zero.
2. Podemos realizar adi√ß√£o vetorial e multiplica√ß√£o escalar.

Existem dois fatos sobre os subespa√ßos:

1. Qualquer linha que passa $(0, 0)$ em $\mathbb{R}^2$ √© um subespa√ßo de $\mathbb{R}^2$.
2. Qualquer plano que passa $(0, 0, 0)$ em $\mathbb{R}^3$ √© um subespa√ßo de $\mathbb{R}^3$.

A seguir vamos visualizar o subespa√ßo.

## O Que √© Subespa√ßo Vetorial?

Um **subespa√ßo vetorial** √© um conceito fundamental da **√Ålgebra Linear**. Trata-se de um **conjunto de vetores** dentro de um espa√ßo vetorial maior que satisfaz algumas propriedades espec√≠ficas. Para que um conjunto seja considerado um subespa√ßo vetorial, ele deve atender √†s **tr√™s condi√ß√µes abaixo**:

### 1. Fechamento sob Adi√ß√£o
Se voc√™ pegar **dois vetores** quaisquer dentro do subespa√ßo e **soma-los**, o vetor resultante **tamb√©m deve pertencer** ao subespa√ßo.

> Formalmente:  
Se **ùë¢** e **ùë£** pertencem ao subespa√ßo, ent√£o **ùë¢ + ùë£** tamb√©m pertence.

### 2. Fechamento sob Multiplica√ß√£o Escalar
Se voc√™ pegar um vetor do subespa√ßo e **multiplic√°-lo por um escalar** (n√∫mero real ou complexo), o vetor resultante **deve continuar no subespa√ßo**.

> Formalmente:  
Se **ùë£** pertence ao subespa√ßo e **ùëê** √© um escalar, ent√£o **ùëêùë£** tamb√©m pertence.

### 3. Cont√©m o Vetor Nulo
Todo subespa√ßo vetorial **deve conter o vetor nulo** (ou seja, o vetor com todas as componentes iguais a zero).

> Isso ocorre porque:  
- Adicionar o vetor nulo a qualquer vetor n√£o altera o vetor.  
- Multiplicar qualquer vetor pelo escalar zero resulta no vetor nulo.


### Exemplo em ‚Ñù¬≥ (espa√ßo tridimensional)
Considere o espa√ßo vetorial ‚Ñù¬≥ (todos os vetores tridimensionais com entradas reais):

- Uma **linha** que **passa pela origem** √© um subespa√ßo vetorial.
- Um **plano** que **passa pela origem** tamb√©m √© um subespa√ßo vetorial.

No entanto:
- Uma linha ou plano que **n√£o passa pela origem** **n√£o** √© subespa√ßo, pois **n√£o cont√©m o vetor nulo**.


### Import√¢ncia
Subespa√ßos vetoriais ajudam a:
- Compreender a **estrutura interna** de um espa√ßo vetorial.
- **Decompor** espa√ßos em partes menores e mais manej√°veis.
- Estudar **transforma√ß√µes lineares** e **solu√ß√µes de sistemas lineares** com mais clareza.


### Aplica√ß√£o do Conceito de Subespa√ßo Vetorial em Data Science e Machine Learning

O conceito de subespa√ßo vetorial √© fundamental em Data Science e Machine Learning, desempenhando um papel importante em v√°rias t√©cnicas e algoritmos. Abaixo est√£o algumas de suas principais aplica√ß√µes:

**Clustering e Classifica√ß√£o:**

Algoritmos como **K-means** e **Support Vector Machines (SVM)** operam em espa√ßos vetoriais. No caso do SVM, por exemplo, um **hiperplano** (subespa√ßo vetorial) √© utilizado para separar as classes. A escolha apropriada desse hiperplano impacta diretamente o desempenho do classificador.

**Processamento de Linguagem Natural (PLN):**

T√©cnicas como **an√°lise sem√¢ntica latente (LSA)** usam subespa√ßos vetoriais para representar significados de palavras e documentos. Essa proje√ß√£o permite capturar:

**An√°lise de S√©ries Temporais**

Na modelagem de s√©ries temporais, subespa√ßos vetoriais ajudam a identificar:
- Tend√™ncias,

**Redes Neurais**

Cada camada de uma **rede neural profunda** pode ser vista como uma transforma√ß√£o de um espa√ßo vetorial para outro. A aprendizagem consiste em encontrar subespa√ßos que melhor representem os dados, tornando a rede capaz de extrair caracter√≠sticas relevantes para a tarefa final.

### Visualiza√ß√£o do Subespa√ßo vetorial em $\mathbb{R}^3$

Considere um intervalo de dois vetores $u = (1,-2,1)^S$ e $v=(2,1,2)^T$. O intervalo de $(u,v)$ √© um subespa√ßo de $\mathbb{R}^3$, onde $s$ e $t$ s√£o os escalares dos vetores.

$$
\left[
\begin{matrix}
x\\
y\\
z
\end{matrix}
\right]=
s\left[
\begin{matrix}
1\\
-2\\
1
\end{matrix}
\right]+
t\left[
\begin{matrix}
2\\
1\\
2
\end{matrix}
\right]=
\left[
\begin{matrix}
s+2t\\
-2s+t\\
s+2t
\end{matrix}
\right]
$$

Vamos tamb√©m definir um plano que n√£o √© um subespa√ßo adicionando $2$ na terceira equa√ß√£o, ou seja, $z= s+2t+2$.

E ent√£o criamos o plot.

```python
# Cria a figura
fig = plt.figure(figsize = (8, 8))

# Proje√ß√£o 3D
ax = fig.add_subplot(111, projection = '3d')

# Valores de s e t
s = np.linspace(-1, 1, 10)
t = np.linspace(-1, 1, 10)

# Grid
S, T = np.meshgrid(s, t)

# Valores de x, y e z
X = S + 2 * T
Y = -2 * S + T
Z = S + 2 * T

# Plot
ax.plot_surface(X, Y, Z, alpha = .9, cmap = plt.cm.coolwarm)

# Isso n√£o √© mais um subespa√ßo
Z2 = S + 2 * T + 2 

# Plot
ax.plot_surface(X, Y, Z2, alpha = .6, cmap = plt.cm.jet)

# Scatter
ax.scatter(0,0,0, s = 200, color = 'red')

# Texto
ax.text(0, 0, 0, '$(0,0,0)$', size = 18, zorder = 5)

# T√≠tulo
ax.set_title('Visualiza√ß√£o de Subespa√ßo em $\mathbb{R}^3$', x = .5, y = 1.1, size = 20)

# Labels
ax.set_xlabel('Eixo x', size = 18)
ax.set_ylabel('Eixo y', size = 18)
ax.set_zlabel('Eixo z', size = 18)
plt.show()
```

### Opera√ß√µes com Tensores em $\mathbb{R}^4$

Para criar dois tensores de quatro dimens√µes e realizar uma opera√ß√£o de produto escalar (dot product) entre eles, devemos primeiro esclarecer como o produto escalar √© definido para tensores de ordem superior. Em tensores de dimens√µes superiores, o produto escalar pode n√£o ser t√£o direto como em vetores ou matrizes.

No entanto, uma abordagem comum √© tratar os tensores de ordem superior como cole√ß√µes de matrizes ou vetores e realizar opera√ß√µes de produto escalar entre essas matrizes ou vetores correspondentes. Para tensores 4D, voc√™ pode visualiz√°-los como uma cole√ß√£o de matrizes 3D, onde voc√™ pode calcular o produto escalar entre matrizes correspondentes.

Vamos criar dois tensores 4D aleat√≥rios e realizar uma opera√ß√£o que se assemelha a um "produto escalar" entre eles, tratando-os como cole√ß√µes de matrizes 3D. Neste caso, vamos calcular a soma dos produtos escalares das matrizes correspondentes em cada tensor.

O resultado da opera√ß√£o que se assemelha a um "produto escalar" entre os dois tensores 4D aleat√≥rios √© aproximadamente 23.48

Neste c√°lculo, tratamos cada tensor como uma cole√ß√£o de matrizes 3D e calculamos a soma dos produtos escalares de todas as matrizes correspondentes entre os dois tensores.

√â importante notar que essa abordagem √© uma das v√°rias maneiras de se realizar opera√ß√µes entre tensores de alta ordem, e a defini√ß√£o espec√≠fica do "produto escalar" pode variar dependendo do contexto e da aplica√ß√£o. 

```python
import numpy as np

## Criando dois tensores 4D 
tensor1 = np.random.rand(3, 3, 3, 3)  # Tensor com dimens√µes 3x3x3x3
tensor2 = np.random.rand(3, 3, 3, 3)  # Tensor com dimens√µes 3x3x3x3

tensor1.shape
```

```python
print(tensor1)
```

```python
# Realizando a opera√ß√£o "produto escalar" entre os tensores
# Tratando cada tensor como uma cole√ß√£o de matrizes 3D
produto_escalar = 0
for i in range(tensor1.shape[0]):
    for j in range(tensor1.shape[1]):
        produto_escalar += np.tensordot(tensor1[i, j], tensor2[i, j])
```

```python
print(produto_escalar)
```

## √Ålgebra Linear no Espa√ßo Vetorial e Subespa√ßo Vetorial

**Quais pa√≠ses tinham a Economia mais parecida com o Brasil, com base em alguns indicadores econ√¥micos, no ano de 2019 antes da pandemia do Covid-19?**

Indicadores econ√¥micos a considerar:

- Agriculture, forestry, and fishing, value added (current US$) = "NV.AGR.TOTL.CD"

```python
# Define os dados para download do World Bank (BancoMundial)
codigos = ["NE.EXP.GNFS.CD", "NE.IMP.GNFS.CD", "NV.AGR.TOTL.CD", "NY.GDP.MKTP.CD", "NE.RSB.GNFS.CD"]

dados = wb.download(country = "all", indicator = codigos, start = 2019, end = 2019).reset_index()
```

```python
dados.head()
```

```python
paises = wb.get_countries()
paises.head()
```

```python
paises2 = paises[paises["region"] != "Aggregates"].name
paises2
```

```python
dados2 = dados[dados["country"].isin(paises2)].dropna()
dados2.head()
```

```python
dados2.shape
```

```python
# Extrai o vetor de dados para cada pa√≠s
# Cria um dicion√°rio para armazenar os vetores
vetores = {}
for rowid, row in dados2.iterrows():
    vetores[row["country"]] = row[codigos].values

vetores
```

```python
euclid = {}
cosine = {}

target = "Brazil"
```

A norma ou m√≥dulo de um vetor √© o comprimento desse vetor, que pode ser calculado por meio da dist√¢ncia de seu ponto final at√© a origem. O m√≥dulo de um n√∫mero real ‚Äúa‚Äù √© um n√∫mero real que indica o tamanho do segmento de reta das extremidades ‚Äú0‚Äù e ‚Äúa‚Äù ou a dist√¢ncia do ponto ‚Äúa‚Äù at√© o ponto ‚Äú0‚Äù na reta num√©rica.

A norma de um vetor, tamb√©m conhecida como m√≥dulo de um vetor, √© obtida por meio do c√°lculo do comprimento desse vetor.

Em rela√ß√£o aos vetores, eles s√£o objetos matem√°ticos definidos em qualquer tipo de espa√ßo, seja ele uma reta, seja um plano ou espa√ßos com muitas dimens√µes. Al√©m disso, s√£o segmentos de reta orientados criados para descrever movimentos retil√≠neos e possuem marca√ß√£o de dire√ß√£o, sentido e intensidade. Como se trata de segmentos de reta antes de tudo, √© poss√≠vel medir seus comprimentos utilizando c√°lculos que envolvem dist√¢ncia entre dois pontos.

```python
# Loop para calcular a norma dos vetores
for country in vetores:
    
    # Target
    vecA = vetores[target]
    
    # Pa√≠s
    vecB = vetores[country]
    
    # Calcula a dist√¢ncia euclidiana com a opera√ß√£o entre os vetores de dados
    dist = np.linalg.norm(vecA - vecB)
    
    # Calcula a dist√¢ncia cosine (cosseno) com a opera√ß√£o entre os vetores de dados
    cos = np.dot(vecA, vecB) / (np.linalg.norm(vecA) * np.linalg.norm(vecB))
    
    # Carrega os dicion√°rios
    euclid[country] = dist    
    cosine[country] = 1-cos  
```

```python
import pandas as pd

df_distance = pd.DataFrame({"euclid": euclid, "cos": cosine})
```

```python
print("Pa√≠s mais pr√≥ximo do Brasil de acordo com a Dist√¢ncia Euclidiana:")
print(df_distance.sort_values(by = "euclid").head())
```

```python
print("Pa√≠s mais pr√≥ximo do Brasil de acordo com a Dist√¢ncia Cosine (Cosseno):")
print(df_distance.sort_values(by = "cos").head())
```

A dist√¢ncia euclidiana e a dist√¢ncia cosseno s√£o m√©tricas amplamente usadas em machine learning para medir a diferen√ßa entre vetores, mas com prop√≥sitos distintos. A dist√¢ncia euclidiana mede o comprimento do caminho direto entre dois pontos no espa√ßo, sendo √∫til quando a magnitude dos vetores √© relevante, como em dados com escalas padronizadas ou posi√ß√µes f√≠sicas.

J√° a dist√¢ncia cosseno foca no √¢ngulo entre os vetores, desconsiderando suas magnitudes, o que a torna ideal para comparar a dire√ß√£o entre vetores,
A dist√¢ncia cosseno √©  vantajosa quando lidamos com dados que t√™m escalas diferentes, porque ela mede apenas o √¢ngulo entre os vetores, ignorando suas magnitudes. Isso significa que ela avalia a dire√ß√£o dos vetores, e n√£o o tamanho. 

## Exerc√≠cios

Exerc√≠cio 1 - C√°lculo de Dist√¢ncia Euclidiana: Escreva uma fun√ß√£o em Python que receba dois vetores (listas de n√∫meros) de mesma dimens√£o e retorne a dist√¢ncia euclidiana entre eles.

```python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

def distancia_euclidiana(v1, v2):
    return np.sqrt(np.sum((v1 - v2) ** 2))

print("Dist√¢ncia Euclidiana:", distancia_euclidiana(v1, v2))
```


Exerc√≠cio 2 - Dist√¢ncia Euclidiana com Valida√ß√£o de Dados: Modifique a fun√ß√£o do Exerc√≠cio 1 para validar se os dois vetores t√™m o mesmo tamanho. Se n√£o tiverem, a fun√ß√£o deve retornar um erro.

```python
def distancia_euclidiana(v1, v2):
    if len(v1) != len(v2):
        raise ValueError("Os vetores devem ter o mesmo tamanho.")
    return np.sqrt(np.sum((v1 - v2) ** 2))

v1 = np.array([1, 2, 3, 4])
v2 = np.array([4, 5, 6])

try:
    print("Dist√¢ncia Euclidiana:", distancia_euclidiana(v1, v2))
except ValueError as e:
    print("Erro:", e)
```


Exerc√≠cio 3 - Dist√¢ncia Euclidiana em N-Dimens√µes:  Generalize  a  fun√ß√£o  do  Exerc√≠cio  1  para  que  ela  possa  calcular  a  dist√¢ncia euclidiana entre dois vetores em um espa√ßo de qualquer n√∫mero de dimens√µes.

```python
def distancia_euclidiana(v1, v2):
    if len(v1) != len(v2):
        raise ValueError("Os vetores devem ter o mesmo tamanho.")
    return sum((a - b)**2 for a,b in zip(v1, v2))**0.5

v1 = np.array([1, 2, 3, 4])
v2 = np.array([4, 5, 6, 5])
try:
    print("Dist√¢ncia Euclidiana:", distancia_euclidiana(v1, v2))
except ValueError as e:
    print("Erro:", e)
```


Exerc√≠cio 4 - Comparando Dist√¢ncias:  Escreva  um  script  que  receba  tr√™s  vetores  de  mesma  dimens√£o.  Calcule  a dist√¢ncia  euclidiana  entre  o  primeiro  e  os  outros  dois  vetores.  Imprima  qual  dos  dois vetores est√° mais pr√≥ximo do primeiro, baseado na dist√¢ncia euclidiana.


```python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
v3 = np.array([7, 8, 9])

def comparar_distancias(v1, v2, v3):
    dist_v1_v2 = distancia_euclidiana(v1, v2)
    dist_v1_v3 = distancia_euclidiana(v1, v3)
    
    if dist_v1_v2 < dist_v1_v3:
        return "O vetor 2 est√° mais pr√≥ximo do vetor 1."
    else:
        return "O vetor 3 est√° mais pr√≥ximo do vetor 1."
    
print(comparar_distancias(v1, v2, v3))
```
